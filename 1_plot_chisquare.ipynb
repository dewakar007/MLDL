```python
# Easy code for Best-Fit Tests on Data Distribution
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

# 1) Download dataset
data = sns.load_dataset('iris')['sepal_length']

# 2) Identify data type
print("Data Type:", type(data[0]))

# 3) Plot data
plt.figure(figsize=(10,3))
plt.subplot(1,3,1); plt.hist(data); plt.title("Histogram")
plt.subplot(1,3,2); stats.probplot(data, dist="norm", plot=plt)
plt.subplot(1,3,3); plt.boxplot(data); plt.title("Boxplot")
plt.show()

# 4) Fit data to normal distribution
mu, sigma = stats.norm.fit(data)
print(f"Mean={mu:.2f}, Std={sigma:.2f}")

# 5) Apply Goodness-of-Fit Tests
chi_stat, chi_p = stats.chisquare(data)
ks_stat, ks_p = stats.kstest(data, 'norm', args=(mu, sigma))
sh_stat, sh_p = stats.shapiro(data)

# 6) Compare results
print("\nGoodness-of-Fit Results:")
print(f"Chi-square test: Stat={chi_stat:.3f}, p={chi_p:.3f}")
print(f"K-S test: Stat={ks_stat:.3f}, p={ks_p:.3f}")
print(f"Shapiro-Wilk test: Stat={sh_stat:.3f}, p={sh_p:.3f}")
```
