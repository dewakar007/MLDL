```python
# CNN using AlexNet with real output
import torch
import torchvision.models as models
import torch.nn as nn
import torch.nn.functional as F

# Load pre-trained AlexNet
model = models.alexnet(pretrained=True)

# Modify for 10 classes (example)
model.classifier[6] = nn.Linear(4096, 10)

# Dummy input image (like a real one)
x = torch.randn(1, 3, 224, 224)

# Forward pass
output = model(x)

# Convert to predicted class
pred = torch.argmax(F.softmax(output, dim=1))

print("Output probabilities:\n", F.softmax(output, dim=1))
print("Predicted class:", pred.item())
```
